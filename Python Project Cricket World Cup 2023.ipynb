```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Team': ['India', 'Pakistan', 'England', 'Australia', 'New Zealand', 'South Africa', 'Sri Lanka', 'Bangladesh', 'West Indies', 'Afghanistan'], 'Matches Played': [9, 8, 7, 7, 7, 6, 6, 6, 5, 5]})

plt.pie(df['Matches Played'], labels=df['Team'], autopct="%1.1f%%")
plt.title("Pie Chart of Matches Played by Team in the 2023 Cricket World Cup")
plt.show()

plt.figure(figsize=(13, 8))
plt.bar(df['Team'], df['Matches Played'])
plt.xlabel("Team")
plt.ylabel("Matches Played")
plt.title("Bar Graph of Matches Played by Team in the 2023 Cricket World Cup")
plt.show()
```

```python
import matplotlib.pyplot as plt

teams = ["Australia", "India", "England", "South Africa", "New Zealand"]
top_wicket_takers = [("Mitchell Starc", 25), ("Jasprit Bumrah", 22), ("Jofra Archer", 21), ("Lungi Ngidi", 19), ("Tim Southee", 18)]

player_dict = {}
for team, player_wickets in zip(teams, top_wicket_takers):
    player_dict[team] = player_wickets[0]

players = list(player_dict.values())
wickets = [player_wickets[1] for player_wickets in top_wicket_takers]

plt.figure(figsize=(8, 4))
plt.bar(teams, wickets, color=['gold', 'blue', 'green', 'red', 'purple'])
plt.xlabel("Teams")
plt.ylabel("Wickets Taken")
plt.title("Top Wicket-Takers in Cricket World Cup 2023 (by Team)")
plt.xticks(rotation=45, ha="right")  

for i, (team, player) in enumerate(zip(teams, players)):
    plt.text(i, wickets[i] + 0.5, player, ha='center', va='bottom', fontsize=12) 

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
s
world_cup_years = ['1975', '1979', '1983', '1987', '1992', '1996', '1999', '2003', '2007', '2011', '2015', '2019']
champions = ['West Indies', 'West Indies', 'India', 'Australia', 'Pakistan', 'Sri Lanka', 'Australia', 'Australia', 'Australia', 'India', 'Australia', 'England']

champion_counts = {}
for team in champions:
    if team in champion_counts:
        champion_counts[team] += 1
    else:
        champion_counts[team] = 1

plt.figure(figsize=(8, 4))
plt.bar(champion_counts.keys(), champion_counts.values(), color='gold')
plt.title('Number of ICC Cricket World Cup Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')

plt.tight_layout()
plt.show()
```
